```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#use SN to determine number of total players without duplicates
total_players= len(purchase_data["SN"].value_counts())

#put into data frame to report result
player_count = pd.DataFrame({"Total_Players":[total_players]})
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of unique items count
U_items = len(purchase_data["Item Name"].unique())

#Average price calcuation and formatting
Average_Price = "${:.2f}".format(purchase_data["Price"].mean())

#Number of purchases
Number_of_Purchases = len(purchase_data["Purchase ID"])

#Sum of Revenue
Total_Revenue = "${:,.2f}".format(purchase_data["Price"].sum())

#create dataframe summary
Purchasing_Analysis = pd.DataFrame({
    "Number of Unique items":[U_items],
    "Average Price":[Average_Price],
    "Number of Purchases":[Number_of_Purchases],
    "Total Revenue": [Total_Revenue]
})
Purchasing_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins and labels
age_bins = [0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,200]
age_brackets = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#use pd.cut function to catergorise the existing players
purchase_data["Age Brackets"] = pd.cut(purchase_data["Age"],age_bins, labels=age_brackets,include_lowest=False)

age_grp_df = purchase_data.groupby("Age Brackets")

#Purchase count
age_purchase_count = age_grp_df["SN"].nunique()

#Percentage of players
age_percentage = age_purchase_count/total_players

age_purchasing = pd.DataFrame({
    "Total Count":age_purchase_count,
    "Percentage of Players":age_percentage
})
#format percentage
age_purchasing.loc[:, "Percentage of Players"] = age_purchasing["Percentage of Players"].map('{:.2%}'.format)
age_purchasing
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Purchase count by age
age_grp_count= age_grp_df["Purchase ID"].count()

#Average purchase price
avg_age_grp_purchase = age_grp_df["Price"].mean()

#Average purchase total
avg_age_purchase_total = age_grp_df["Price"].sum()

#Average purchase total per person
avg_age_total_pp = avg_age_purchase_total/ age_purchase_count

#Build data frame
age_purchasing_anyalsis = pd.DataFrame({
    "Purchase Count":age_grp_count,
    "Average Purchase Price":avg_age_grp_purchase,
    "Total Purchase Value":avg_age_purchase_total,
    "Avg Total Purchase per Person":avg_age_total_pp
})
#table formating
age_purchasing_anyalsis.loc[:, "Total Purchase Value"] = age_purchasing_anyalsis["Total Purchase Value"].map('${:,.2f}'.format)
age_purchasing_anyalsis.loc[:, "Average Purchase Price"] = age_purchasing_anyalsis["Average Purchase Price"].map('${:,.2f}'.format)
age_purchasing_anyalsis.loc[:, "Avg Total Purchase per Person"] = age_purchasing_anyalsis["Avg Total Purchase per Person"].map('${:,.2f}'.format)

age_purchasing_anyalsis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by spender
spender_df = purchase_data.groupby("SN")

#purchase count
spender_count = spender_df["Purchase ID"].count()

#Average purchase price
spender_avg_purch_price = spender_df["Price"].mean()

#Total purchase value
spender_total = spender_df["Price"].sum()

#Build data frame
top_spender_df = pd.DataFrame({
    "Purchase Count":spender_count,
    "Average Purchase Price":spender_avg_purch_price,
    "Total Purchase Value":spender_total
})
#sort bale by purchase value
sorted_top_spender_df = top_spender_df.sort_values("Total Purchase Value",ascending=False)

#table formating
sorted_top_spender_df.loc[:, "Total Purchase Value"] = sorted_top_spender_df["Total Purchase Value"].map('${:,.2f}'.format)
sorted_top_spender_df.loc[:, "Average Purchase Price"] = sorted_top_spender_df["Average Purchase Price"].map('${:,.2f}'.format)

sorted_top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Extract Item ID, Item Name and Item Price columns
pop_items = purchase_data[["Item ID", "Item Name", "Price"]]

pop_item_df = pop_items.groupby(["Item ID", "Item Name"])

#count of popular items
pop_item_count = pop_item_df["Item ID"].count()

#Item Price
pop_item_price = pop_item_df["Price"].mean()

# Total purchase value
pop_item_value = pop_item_df["Price"].sum()


#Build data frame
pop_item_analysis = pd.DataFrame({
    "Purchase Count":pop_item_count,
    "Item Price":pop_item_price,
    "Total Purchase Value":pop_item_value
})

pop_item_analysis.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sort bale by purchase value
pop_item_analysis= pop_item_analysis.sort_values("Total Purchase Value",ascending=False)


# table formating
pop_item_analysis.loc[:, "Total Purchase Value"] = pop_item_analysis["Total Purchase Value"].map('${:,.2f}'.format)
pop_item_analysis.loc[:, "Item Price"] = pop_item_analysis["Item Price"].map('${:,.2f}'.format)

pop_item_analysis.head()
```
